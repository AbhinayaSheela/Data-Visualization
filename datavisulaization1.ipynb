```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df= pd.read_csv("C:\practise\day9\dataset.csv")
df
```

```python
#figure size
plt.figure(figsize=(5,5))
#histogram
sns.histplot(df["age"], bins=20, kde=True, color="black", alpha=0.7, edgecolor="pink")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x=df["age"], y=df["trestbps"], color="grey", alpha=0.6, edgecolor="black", s=100, marker='^')
plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x=df["trestbps"], y=df["chol"], color="black", alpha=0.5, edgecolor="black", s=50, marker='o')
plt.title("Blood Pressure vs Cholesterol")
plt.xlabel("Blood Pressure")
plt.ylabel("Cholesterol")
plt.grid(True)
plt.show()
```

```python
#line plot
plt.figure(figsize=(5,5))
sns.lineplot(x=df["age"], y=df["chol"], color="purple", marker="d", linestyle="--", linewidth=1, markersize=8)
plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.grid(True)
plt.show()
```

```python
#box plot
plt.figure(figsize=(7,5))
sns.boxplot(x=df["sex"], y=df["chol"], color="black", fliersize=5, linewidth=2, saturation=0.8, boxprops=dict(facecolor="hotpink"))
plt.title("Cholesterol by Sex")
plt.xlabel("Sex (2=Female, 2=Male)")
plt.ylabel("Cholesterol")
plt.grid(True)
plt.show()
```

```python
#pie chart with custom colors and edges
plt.figure(figsize=(6,6))
labels = df["sex"].value_counts().index
sizes = df["sex"].value_counts().values
colors = ["brown", "gold"]
plt.pie(sizes, labels=["Male", "Female"], colors=colors, autopct="%1.1f%%", startangle=140, wedgeprops=dict(edgecolor="black", linewidth=1),shadow=True, explode=(0.1, 0))
plt.title("Sex Distribution")
plt.axis("equal")
plt.show()
```

```python
#barchart with interesting presentation
plt.figure(figsize=(7,5))
sns.barplot(x=df["sex"], y=df["chol"], color="cyan", edgecolor="black", linewidth=1, saturation=0.8)
plt.title("Average Cholesterol by Sex")
plt.xlabel("Sex (2=Female, 2=Male)")
plt.ylabel("Average Cholesterol")
plt.grid(True)
plt.show()
```

```python
#bar chart with many categories
plt.figure(figsize=(8,6))
sns.barplot(x=df["cp"], y=df["chol"], color="beige", edgecolor="black", linewidth=0.2, saturation=0.8)
plt.title("Average Cholesterol by Chest Pain Type")
plt.xlabel("Chest Pain Type (0-3)")
plt.ylabel("Average Cholesterol")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Calculate average cholesterol by chest pain type
avg_chol = df.groupby("cp")["chol"].mean()

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

# X positions
x = np.arange(len(avg_chol))
y = np.zeros(len(avg_chol))
z = np.zeros(len(avg_chol))

dx = np.ones(len(avg_chol)) * 0.5
dy = np.ones(len(avg_chol)) * 0.5
dz = avg_chol.values

# Weird neon colors
colors = ["#ff00ff", "#00ffff", "#39ff14", "#ff1493"]

ax.bar3d(x, y, z, dx, dy, dz, color=colors, shade=True)

# Weird styling
ax.set_xlabel("Chest Pain Dimension ðŸ˜µ")
ax.set_ylabel("Invisible Axis ðŸ‘»")
ax.set_zlabel("Cholesterol Height ðŸ’‰")

ax.set_xticks(x)
ax.set_xticklabels(["Silent ðŸ˜¶", "Mild ðŸ˜¬", "Serious ðŸ˜µ", "Emergency ðŸš‘"])

ax.set_facecolor("black")
fig.patch.set_facecolor("black")

plt.title("3D Cholesterol Multiverse ðŸŒŒ", color="white")
# Floating spheres
for i in range(10):
    ax.scatter(np.random.uniform(0,3),
               np.random.uniform(0,1),
               np.random.uniform(0,300),
               color=np.random.rand(3,),
               s=200)

plt.show()
```
