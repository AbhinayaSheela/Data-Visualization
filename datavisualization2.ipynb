```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\practise\day9\Practice-Datasets .csv")
```

```python
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
df['City'].value_counts().plot.bar()
plt.title("Customers by City")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(y=df['Monthly_Income'])
plt.title("Monthly Income Distribution")
plt.show()
```

```python
plt.hist(df['Spending_Score'], bins=10)
plt.title("Spending Score Distribution")
plt.xlabel("Spending Score")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

df = df.sort_values('Purchase_Date')

plt.plot(df['Purchase_Date'], df['Purchase_Amount'])
plt.xticks(rotation=45)
plt.title("Purchase Amount Trend")
plt.show()
```

```python
df.groupby('Gender')['Monthly_Income'].mean().plot.bar()
plt.title("Average Monthly Income by Gender")
plt.show()
```

```python
sns.boxplot(x='Product_Category', y='Purchase_Amount', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
df['Product_Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Product Category Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.hist(df['Purchase_Amount'], bins=15)
plt.title("Purchase Amount Distribution")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby('City')['Spending_Score'].mean().plot.bar()
plt.title("Average Spending Score by City")
plt.show()
```

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

monthly = df.resample('M', on='Purchase_Date')['Purchase_Amount'].sum()

plt.plot(monthly.index, monthly.values)
plt.xticks(rotation=45)
plt.title("Total Purchase Amount per Month")
plt.show()
```

```python
sns.boxplot(x='City', y='Monthly_Income', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
for gender in df['Gender'].unique():
    plt.hist(df[df['Gender'] == gender]['Age'], bins=10, alpha=0.5, label=gender)

plt.legend()
plt.title("Age Distribution by Gender")
plt.show()
```

```python
df.groupby('Product_Category')['Purchase_Amount'].sum().plot.bar()
plt.title("Total Purchase Amount by Product Category")
plt.show()
```

```python
plt.scatter(df['Monthly_Income'], df['Spending_Score'])
plt.xlabel("Monthly Income")
plt.ylabel("Spending Score")
plt.title("Income vs Spending Score")
plt.show()
```

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

daily = df.resample('D', on='Purchase_Date')['Purchase_Amount'].sum()

plt.plot(daily.index, daily.values)
plt.xticks(rotation=45)
plt.title("Daily Purchase Amount Variation")
plt.show()
```

```python
sns.boxplot(x='Gender', y='Spending_Score', data=df)
plt.show()
```

```python
pd.crosstab(df['Product_Category'], df['Gender']).plot(kind='bar')
plt.title("Customer Count by Category and Gender")
plt.show()
```

```python
plt.scatter(df['Age'], df['Purchase_Amount'])
plt.xlabel("Age")
plt.ylabel("Purchase Amount")
plt.title("Age vs Purchase Amount")
plt.show()
```
